```python
# ans 1

#getting the values

df.target.value_counts()
```

```python
#plot bar chart
df.target.value_counts().plot(kind='bar',color=["purple", "Orange"])
plt.title("Heart disease values")
plt.xlabel("1= heart disease,0= no heart disease")
plt.ylabel("amount")
plt.show()
```

```python
#plot pie chart

df.target.value_counts().plot(kind='pie',figsize=(8,6))
plt.legend(['Disease',"no disease"]);


#0 represent  'female'
# 1 represent 'male'

```

```python
#target column part

# Q 2  lets check how many male and female are there in dataset

df.sex.value_counts()
```

```python
#plot in graph

df.sex.value_counts().plot(kind='pie',figsize=(5,6))
plt.title("male female ratio")
plt.legend(["male","female"])
```

```python
#Q3  people of which sex has more heart disease

pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x="target",data=df,hue="sex")
plt.title("heart disease frequency for sex")
plt.xlabel("0= no heart disease,1= heart disease")
```

```python
## hear no of male is more then double in our dataset then female.
## more then '45% male' has heart disease and '75% female' has heart disease
```

```python
# Q4  people of which sex has which type of chest pain  most?
 #counting diffrent type of chest pain

df.cp.value_counts()
```

```python
#plot it

df.cp.value_counts().plot(kind='bar',color=["salmon","lightskyblue","khaki","springgreen"])
plt.title("chest pain type vs count")
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind='bar',color =['coral','red','blue','purple'])
plt.title("type of chest pain for sex")
plt.xlabel("0=female, 1=male")
```

```python
# most of people who has 'type 0' chest pain has less chance of heart disease.
# and we see the opposite for other types,

# Q 5  now lets look at age column
#create a distributed plt with normal distribution curves

sns.displot(x='age',data=df,bins=30,kde=True,)
```
